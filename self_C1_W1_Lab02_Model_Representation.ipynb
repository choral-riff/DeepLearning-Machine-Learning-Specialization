```python
#x_train is the input variable (size in 1000 square fts)
#y_train is the target (price in 1000s of dollars)
x_train = np.array([1.0,2.0])
y_train = np.array([300,500])
print(f"x_train = {x_train}")
print(f"y_train ={y_train}")
```

```python
#Training examples x_i and y_i
i=0

x_i=x_train[i]
y_i=y_train[i]
print(f"(x^({i}),y^({i}))=({x_i},{y_i})")

i=1

x_i=x_train[i]
y_i=y_train[i]
print(f"(x^({i}),y^({i}))=({x_i},{y_i})")
```

```python
#Plotting the data points
plt.scatter(x_train, y_train, marker='x', c='r')
#setting the title
plt.title("Housing Prices")
#settig the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
#setting the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.show()
```

Model Function:

for univariate linear regression, function is represented by:

f(x) = w*x + b

```python
w = 200
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
#f is to be computed for several i = 1....m
#making a function compute_model_output
def compute_model_output(x, w, b):
  m = len(x)
  f_wb = np.zeros(m)
  for i in range(m):
    f_wb[i] = w*x[i] + b

  return f_wb
```

```python
#calling compute_model_output method and plotting the output
tmp_f_wb = compute_model_output(x_train, w, b)
#plotting our model prediction
plt.plot(x_train, tmp_f_wb, c='b', label='Our Prediction')

#Plot the actual data points
plt.plot(x_train, y_train, marker='x', c='r', label='Actual Values')

#plot title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

Our prediction matches the actual datatrend exactly.
